```python
import pandas as pd

df = pd.read_csv("../data/cleaned_hospital_data.csv")
```

```python
import matplotlib.pyplot as plt

status_cost = (
    df.groupby("status")["treatment_cost"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure()
status_cost.plot(kind="bar")
plt.title("Average Treatment Cost by Appointment Status")
plt.xlabel("Appointment Status")
plt.ylabel("Average Treatment Cost")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
“This analysis shows how treatment costs vary by appointment status.
Completed appointments generate the highest cost, while cancellations and no-shows indicate potential revenue loss.
This insight can help hospitals improve scheduling and reduce operational inefficiencies.”
```

```python
# Hospital Appointment System – Business Insights

This notebook summarizes key insights derived from exploratory data analysis
and provides recommendations to improve hospital appointment operations.
```

```python
## Data Overview

The dataset contains hospital appointment records including:
- Appointment dates
- Doctor IDs
- Patient visit reasons
- Appointment status (Completed / Cancelled / No-show)

The analysis focuses on identifying operational patterns and improvement areas.
```

```python
## Key Insights

### 1. Appointment Status Trends
- A significant portion of appointments are either cancelled or result in no-shows.
- This indicates inefficiencies in scheduling and patient commitment.

### 2. High-Frequency Visit Reasons
- Certain reasons for visit appear repeatedly among top appointments.
- These may require specialized staff allocation or dedicated time slots.

### 3. Doctor Workload Imbalance
- A small number of doctors handle a disproportionately high number of appointments.
- This could lead to burnout and longer patient wait times.

### 4. Seasonal Appointment Patterns
- Appointment volume varies by month, showing seasonal demand trends.
- Resource planning can be optimized based on these trends.
```

```python
## Business Recommendations

1. Implement automated appointment reminders (SMS / Email)  
   → To reduce cancellations and no-shows.

2. Introduce dynamic scheduling  
   → Allocate more slots to high-demand visit reasons.

3. Balance doctor workload  
   → Redistribute appointments or hire temporary staff during peak periods.

4. Plan resources seasonally  
   → Adjust staffing levels based on monthly appointment trends.
```

```python
## Conclusion

This analysis highlights operational inefficiencies and demand patterns
within the hospital appointment system.

By applying data-driven scheduling, workload balancing, and reminder systems,
the hospital can improve patient experience and overall efficiency.
```
