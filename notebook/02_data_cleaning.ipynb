```python
import pandas as pd

df = pd.read_csv("../data/raw_hospital_data.csv", encoding="latin1")
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.isnull().sum()
df = df.dropna()
```

```python
df["treatment_cost"] = (
    df["treatment_cost"]
    .astype(str)
    .str.replace("[₹,]", "", regex=True)
    .astype(float)
)
```
